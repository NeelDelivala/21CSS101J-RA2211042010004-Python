```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

x = []
y = []
plt.rcParams["figure.autolayout"] = True
header = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']
df = pd.read_csv('/content/california_housing_test2.csv', names=header)

plt.ylabel('Median Income')
plt.xlabel('Median House Value')
plt.title('California Housing Test')
plt.plot(df.housing_median_age,df.median_income)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df.housing_median_age, df.median_income)
plt.xlabel('Housing Median Age')
plt.ylabel('Median Income')
plt.title('Bar Chart: Median Income by Housing Median Age')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df.population, df.median_house_value)
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.title('Scatter Plot: Population vs Median House Value')
plt.show()
```

```python
plt.figure(figsize=(8,5))

df['total_bedrooms'] = pd.to_numeric(df['total_bedrooms'], errors='coerce')
plt.boxplot(df['total_bedrooms'].dropna(), patch_artist=True, notch=True,
            boxprops=dict(facecolor='lightblue'))

plt.ylabel('Total Bedrooms')
plt.title('Box Plot: Distribution of Total Bedrooms')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.fill_between(df.index, df.total_rooms, color="skyblue", alpha=0.4)
plt.plot(df.total_rooms, color="Slateblue", alpha=0.6)
plt.xlabel('Index')
plt.ylabel('Total Rooms')
plt.title('Area Chart: Total Rooms by Index')
plt.show()
```

```python
plt.figure(figsize=(13,5))
plt.hist(df['housing_median_age'].dropna(), bins=20, color='teal', edgecolor='black', alpha=0.7)
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')
plt.title('Histogram: Distribution of Housing Median Age')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['median_income'] = pd.to_numeric(df['median_income'], errors='coerce')
df['total_rooms'] = pd.to_numeric(df['total_rooms'], errors='coerce')

df_cleaned = df.dropna(subset=['median_income', 'total_rooms'])

income_bins = pd.cut(df_cleaned['median_income'], bins=4)
room_groups = df_cleaned.groupby(income_bins)['total_rooms'].sum()

plt.pie(room_groups, labels=room_groups.index, autopct='%1.1f%%', startangle=140,
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])

plt.title('Pie Chart: Proportion of Total Rooms by Binned Median Income')
plt.show()
```
